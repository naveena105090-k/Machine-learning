# UNIVARIENT ANALYSIS

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

titanic.head()
```

### BAR CHART

 **What is the distribution of passengers by Gender?**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', data=titanic, palette='pastel', edgecolor='black')
plt.title('Distribution of Passengers by Gender', fontsize=14)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

The majority of passengers were male.

The number of female passengers was comparatively smaller.

**How many survived vs did not survive?**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='survived', data=titanic, palette='Set2', edgecolor='black')
plt.title('Survival Count on the Titanic', fontsize=14)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

A large number of passengers did not survive.

Fewer passengers were rescued.

**What is the distribution of passengers by Travel Class?**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='class', data=titanic, palette='cool', edgecolor='black')
plt.title('Distribution of Passengers by Travel Class', fontsize=14)
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

The Third class had the most passengers.

The First class had the fewest.

**How many passengers were there categorised as man,women,child**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='who', data=titanic, palette='muted', edgecolor='black')
plt.title('Distribution of Passengers by Category (Man, Woman, Child)', fontsize=14)
plt.xlabel('Who')
plt.ylabel('Count')
plt.show()
```

Most passengers were men.

The number of children was very small.

### PIE CHART

**Passenger Class Distribution**

```python
plt.figure(figsize=(6,6))
titanic['class'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('cool'))
plt.title('Passenger Class Distribution (Pie Chart)', fontsize=14)
plt.ylabel('')
plt.show()
```

~55% of passengers were in Third Class.

~25% in First Class.

~20% in Second Class.

### HISTOGRAM

**What is the distribution of Age of passengers?**

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['age'], bins=30, kde=False, color='skyblue', edgecolor='black')
plt.title('Distribution of Passenger Age', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Most passengers were between 20 and 40 years old.

Very few were elderly or young children.

### Kernel Density Estimate (KDE) Plot

**What is the distribution shape of Age?**

```python
plt.figure(figsize=(8,5))
sns.kdeplot(x='age', data=titanic, fill=True, color='salmon')
plt.title('Age Distribution  KDE Plot', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

The curve peaks near 2530 years, indicating that most passengers were young adults.

The distribution is slightly right-skewed, showing more younger than older passengers.

### BOXPLOT

**What is the spread of Age using a Box Plot?**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='age', data=titanic, color='violet')
plt.title('Box Plot of Passenger Age', fontsize=14)
plt.xlabel('Age')
plt.show()
```

Median age is around 28 years.

There are some outliers (older passengers).

The spread (IQR) is roughly 2040 years.

**What is the spread of Fare using a Box Plot?**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='fare', data=titanic, color='gold')
plt.title('Box Plot of Passenger Fare', fontsize=14)
plt.xlabel('Fare')
plt.show()
```

Median fare is relatively low, but a few passengers paid very high fares.

The presence of outliers indicates a wide range of ticket prices across classes.
